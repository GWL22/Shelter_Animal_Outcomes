```python
test_file = './dataset/test.csv'
# test_file = './dataset/train.csv'
with open(test_file, 'rb') as header_file:
    headers = header_file.readline().strip().split(',')
print headers
```

```python
# usecols = headers[2:4] + headers[5:] #for train.csv
usecols = headers[2:] #for test.csv
print usecols
```

```python
# df = pd.read_csv( test_file, usecols=usecols).dropna() # for train.csv
df = pd.read_csv( test_file, usecols=usecols)
```

```python
df['AgeuponOutcome'] = df['AgeuponOutcome'].fillna('1 year')
```

```python
test = df['AgeuponOutcome']
```

```python
for row, index in zip(test, test.index):
    item = row.split(' ')
    item[0] = int(item[0])
#     print index
    if 'year' in item[1]:
        test[index] = item[0] * 365
    elif 'month' in item[1]:
        test[index] = item[0] * 30
    elif 'week' in item[1]:
        test[index] = item[0] * 7
    elif 'day' in item[1]:
        test[index] = item[0]
```

```python
df['AgeOutcomeYear'] = [x/365 for x in df['AgeuponOutcome']]
```

```python
df['AgeOutcomeMonth'] = [x/30 for x in df['AgeuponOutcome']]
```

```python
df['AgeOutcomeWeek'] = [x/7 for x in df['AgeuponOutcome']]
```
